# Bước 6: Phân cụm khách hàng từ Luật kết hợp (Association Rules → Clustering)

Notebook này lấy **kết quả luật kết hợp** (Apriori/FP-Growth) và biến chúng thành **đặc trưng** để phân cụm khách hàng bằng K-Means.

## Ý tưởng cốt lõi
- Mỗi luật có dạng: **Antecedent → Consequent**
- Với mỗi khách hàng, ta kiểm tra: khách đó đã từng mua **đủ antecedents** của luật hay chưa.
- Mỗi luật trở thành một feature (0/1 hoặc có trọng số theo lift/confidence).
- (Tuỳ chọn) Ghép thêm **RFM** để phân cụm ổn định hơn.


## Parameters
Gán tham số để chạy bằng papermill.


```python
# PARAMETERS (for papermill)

# Input
CLEANED_DATA_PATH = "../data/processed/cleaned_uk_data.csv"
RULES_INPUT_PATH = "../data/processed/rules_fpgrowth_filtered.csv"  # hoặc rules_fpgrowth_filtered.csv

# Feature engineering
TOP_K_RULES = 200
SORT_RULES_BY = "lift"      # lift | confidence | support
WEIGHTING = "lift_x_conf"          # none | lift | confidence | support | lift_x_conf
MIN_ANTECEDENT_LEN = 2
USE_RFM = True
RFM_SCALE = True
RULE_SCALE = False

# Clustering
K_MIN = 2
K_MAX = 10
N_CLUSTERS = None            # None => chọn theo silhouette, hoặc đặt số cụ thể (vd 5)
RANDOM_STATE = 42

# Output
OUTPUT_CLUSTER_PATH = "data/processed/customer_clusters_from_rules.csv"

# Visual
PROJECTION_METHOD = "pca"   # pca | svd
PLOT_2D = True
```

## Set up


```python
%load_ext autoreload
%autoreload 2

import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Determine correct project root
cwd = os.getcwd()
if os.path.basename(cwd) == "notebooks":
    project_root = os.path.abspath("..")
else:
    project_root = cwd

src_path = os.path.join(project_root, "src")
if src_path not in sys.path:
    sys.path.append(src_path)

from cluster_library import RuleBasedCustomerClusterer
```

## Load cleaned data & rules


```python
df_clean = pd.read_csv(CLEANED_DATA_PATH, parse_dates=["InvoiceDate"])
print(df_clean.shape)
df_clean.head()
```

```python
clusterer = RuleBasedCustomerClusterer(df_clean=df_clean)
customer_item_bool = clusterer.build_customer_item_matrix(threshold=1)
print('Customer × Item:', customer_item_bool.shape)

rules_df = clusterer.load_rules(
    rules_csv_path=RULES_INPUT_PATH,
    top_k=TOP_K_RULES,
    sort_by=SORT_RULES_BY,
)
print('Rules used:', rules_df.shape)
rules_df.head()
```

## Build features (Rules → Features) + (optional) RFM


```python
X, meta = clusterer.build_final_features(
    weighting=WEIGHTING,
    use_rfm=USE_RFM,
    rfm_scale=RFM_SCALE,
    rule_scale=RULE_SCALE,
    min_antecedent_len=MIN_ANTECEDENT_LEN,
)
print('X shape:', X.shape)
meta.head()
```

## Choose K (silhouette)


```python
sil_df = clusterer.choose_k_by_silhouette(
    X,
    k_min=K_MIN,
    k_max=K_MAX,
    random_state=RANDOM_STATE,
)
sil_df
```

```python
best_k = int(sil_df.loc[0, 'k'])
N_CLUSTERS = 4
k = best_k if N_CLUSTERS is None else int(N_CLUSTERS)
print('Chosen k =', k)
```

## Fit KMeans & save results


```python
labels = clusterer.fit_kmeans(X, n_clusters=k, random_state=RANDOM_STATE)
meta_out = meta.copy()
meta_out['cluster'] = labels

# Lưu
os.makedirs(os.path.dirname(OUTPUT_CLUSTER_PATH), exist_ok=True)
meta_out.to_csv(OUTPUT_CLUSTER_PATH, index=False)
print('Saved:', OUTPUT_CLUSTER_PATH)
meta_out.head()
```

## Quick profiling


```python
profile_cols = ['cluster'] + ([c for c in ['Recency','Frequency','Monetary'] if c in meta_out.columns])
summary = meta_out.groupby('cluster').agg({
    'CustomerID': 'count',
    **{c:'mean' for c in profile_cols if c!='cluster'}
}).rename(columns={'CustomerID':'n_customers'}).sort_values('n_customers', ascending=False)
summary
```

## 2D visualization (PCA/SVD)


```python
if PLOT_2D:
    Z = clusterer.project_2d(X, method=PROJECTION_METHOD, random_state=RANDOM_STATE)
    plt.figure(figsize=(8,6))
    plt.scatter(Z[:,0], Z[:,1], c=labels, s=10)
    plt.title('Customer clusters (2D projection)')
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.tight_layout()
    plt.show()
```

## Phân tích đặc trưng của từng cụm: Top 10 luật kết hợp

Sau khi phân cụm, ta tính trung bình điểm số (weighted score) của từng luật trong mỗi cụm.  
Luật nào có điểm trung bình cao nhất → đặc trưng nhất cho cụm đó.

```python
# Đọc kết quả phân cụm đã lưu
cluster_output_path = OUTPUT_CLUSTER_PATH  # đã định nghĩa ở phần parameters
df_clusters = pd.read_csv(cluster_output_path)

print(f"Đã tải {df_clusters.shape[0]} khách hàng với {df_clusters['cluster'].nunique()} cụm")
df_clusters.head()
```

```python
# Khởi tạo lại clusterer để lấy rules và tên cột đặc trưng
clusterer = RuleBasedCustomerClusterer(df_clean=df_clean)

# Tải lại rules (giống như ở trên)
rules_df = clusterer.load_rules(
    rules_csv_path=RULES_INPUT_PATH,
    top_k=TOP_K_RULES,
    sort_by=SORT_RULES_BY,
)

# Xây dựng lại ma trận đặc trưng X (cùng tham số như lúc clustering)
X, meta = clusterer.build_final_features(
    weighting=WEIGHTING,
    use_rfm=USE_RFM,
    rfm_scale=RFM_SCALE,
    rule_scale=RULE_SCALE,
)

# Danh sách tên cột
feature_columns = []
if USE_RFM:
    feature_columns += ['Recency', 'Frequency', 'Monetary']
feature_columns += rules_df['rule_str'].tolist()  # tên luật dạng "A, B → C"

# DataFrame đặc trưng có index là CustomerID
df_features = pd.DataFrame(X, index=meta['CustomerID'], columns=feature_columns)

# Gắn nhãn cụm vào df_features
df_features = df_features.merge(
    df_clusters[['CustomerID', 'cluster']],
    left_index=True,
    right_on='CustomerID',
    how='left'
)

# Loại bỏ khách hàng không có nhãn cụm (nếu có)
df_features = df_features.dropna(subset=['cluster'])
df_features['cluster'] = df_features['cluster'].astype(int)

print(f"Ma trận đặc trưng cuối cùng: {df_features.shape}")
```

```python
# Chỉ lấy các cột là luật (loại RFM và Cluster)
rule_columns = [col for col in df_features.columns if col not in ['CustomerID','Recency', 'Frequency', 'Monetary','cluster']]

# Group by cụm và tính trung bình cho từng luật
cluster_rule_means = df_features.groupby('cluster')[rule_columns].mean()

# Hiển thị top 10 luật cho từng cụm
n_top = 5

print(f"\n=== TOP {n_top} LUẬT ĐẶC TRƯNG CHO TỪNG CỤM ===\n")

for cluster in sorted(cluster_rule_means.index):
    top_rules = cluster_rule_means.loc[cluster].sort_values(ascending=False).head(n_top)
    
    print(f"🔸 CỤM {cluster} ({len(df_features[df_features['cluster'] == cluster])} khách hàng)")
    for rule, score in top_rules.items():
        print(f"   → {rule}   (điểm tb: {score:.4f})")
    print()
```
